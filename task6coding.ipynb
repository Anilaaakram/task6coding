```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing birth data with columns 'Year', 'Gender', and 'Number of births'

# Step 1: Add a new column "Decade" to the DataFrame
df['Decade'] = (df['Year'] // 10) * 10

# Step 2 & 3: Group the data by decade and gender, then calculate the total number of births for each group
births_by_decade_gender = df.groupby(['Decade', 'Gender'])['Number of births'].sum().reset_index()

# Step 4: Visualize the trend using a line plot
plt.figure(figsize=(10, 6))
for gender in ['Male', 'Female']:
    gender_data = births_by_decade_gender[births_by_decade_gender['Gender'] == gender]
    plt.plot(gender_data['Decade'], gender_data['Number of births'], label=gender)

plt.title('Trend of Male & Female Births Every Decade')
plt.xlabel('Decade')
plt.ylabel('Number of Births')
plt.legend()
plt.grid(True)
plt.show()
```
